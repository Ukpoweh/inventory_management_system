## For Loops

Used with iterables (strings, lists, tuples, sets, and dictionaries)

```python
greeting = "Hello"

print(greeting)
```

```python
#loop over each character in a string
for letter in greeting: #letter is just a placeholder which represents each character in the string
    print(letter)
```

```python
for i in greeting:
    print(f"The letter {i} is present in greeting")
```

```python
#using for loop with lists

student_names = ["John", "Alice", "James"]

print(student_names)
```

```python
#for item in iterable
for name in student_names:
    print(f"Hi, {name}")
    print("Welcome to Techcrush")
```

```python
for name in student_names:
    print(f"Hi, {name}")

print("Welcome to Techcrush")
```

```python
print(student_names)
```

```python
#enumerate function -generator object-returns the index of each item together with the item

list(enumerate(student_names))
```

```python
indexed_list = list(enumerate(student_names))

print(indexed_list)
```

```python
for index, name in indexed_list:
    print(f"The number {index} student is {name}")
```

```python
for index, name in indexed_list:
    position = index + 1
    print(f"The number {position} student is {name}")
```

```python
#using for loops with dictionaries

favourite_fruit = {"Alice":"Pear", "John":"Apple", "James":"Grape"}

print(favourite_fruit)
```

```python
#looping over the keys

for name in favourite_fruit.keys():
    print(name)
```

```python
#looping over the values

for fruit in favourite_fruit.values():
    print(fruit)
```

```python
favourite_fruit.items()
```

```python
for key, value in favourite_fruit.items():
    print(f"{key}'s favourite fruit is {value}")
```

### Range Function

```python
#to create a list of numbers from 1 to 10

list(range(1, 11))
```

```python
list(range(10, 21, 2))
```

```python
for num in range(1, 11):
    print(num)
```

```python
#to create a list of even numbers from 0 to 20

even_numbers = []

for num in range(0, 21):
    if num % 2 == 0:
        even_numbers.append(num)
```

```python
print(even_numbers)
```

```python
odd_numbers = []

for num in range(0, 21):
    if num % 2 == 1:
        odd_numbers.append(num)
```

```python
print(odd_numbers)
```

```python
#to create a list of squares from 1 to 10

squares = []

for num in range(1, 11):
    square = num ** 2
    squares.append(square)

print(squares)
```

### List Comprehension

```python
squares_comp = [num ** 2 for num in range(1, 11)]

print(squares_comp)
```

```python
#for even numbers

even_numbers_comp = [num for num in range(0, 21) if num % 2 == 0]

print(even_numbers_comp)
```

## While Loops

works with conditions

To write a code to count from 1 to 5

```python
counter = 1

while counter <=5:
    print(counter)
    counter += 1
```

To write a code to count from 5 to 1

```python
counter = 5

while counter > 0:
    print(counter)
    counter -= 1
```

using the break keyword

break is used to exit the loop completely, even if the condition is still true

```python
#to count from 1 to 10, but stop when we reach 5

count = 1

while count <= 10:
    #if count == 5:
        #break
    print(count)
    count += 1
```

```python
count = 1

while count <= 10:
    if count == 5:
        break
    print(count)
    count += 1
```

using the continue keyword

continue is used to skip the current iteration and move to the next

```python
#to count from 1 to 10 and skip number 5

count = 0

while count < 10:
    count += 1
    if count == 5:
        continue
    print(count)
```
